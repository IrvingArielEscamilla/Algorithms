```python
#Distribution usin PDF & CDF
counts, bins = np.histogram(df["Age"],bins=10, density= True)
pdf = counts/sum((counts))
cdf = np.cumsum(pdf)
plt.plot(bins[1:], pdf, label="PDF")
plt.plot(bins[1:], cdf, label="CDF")
plt.xlabel("Age")
plt.legend(loc="best")
plt.show()
```

```python
#Distribution using salary income 
counts, bins = np.histogram(df["MonthlyIncome"],bins=10, density= True)
pdf = counts/sum((counts))
cdf = np.cumsum(pdf)
plt.plot(bins[1:], pdf, label="PDF")
plt.plot(bins[1:], cdf, label="CDF")
plt.xlabel("Monthly Income ")
plt.legend(loc="best")
plt.show()
```
