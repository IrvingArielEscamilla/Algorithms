```python
#pytorch mlp for binary classification
from numpy import vstack
from pandas import read_csv
import torch
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
from torch.utils.data import Dataset
from torch.utils.data import DataLoader
from torch import Tensor
from torch.nn import Linear
from torch.nn import ReLU
from torch.nn import Sigmoid
from torch.nn import Module
from torch.optim import SGD
from torch.nn import BCELoss
from torch.nn.init import kaiming_uniform_
from torch.nn.init import xavier_uniform_

#dataset definition
class CSVDataset(Dataset):
    #load the dataset
    def __init__(self,path):
        df = read_csv(path, header=0) ##Add path for dataset
        print(df)
        #store the inputs and outputs
        self.X =df.values[1:, 0:4]
        self.y =df.values[1:, 4:5]
        #print(self.X)
        #print(self.y)
        #sc = standardscaler()
        #self.X = torch.from_numpy(self.X)
        self.X = torch.tensor(self.X, dtype = torch.float32)
        print(self.X)
        #self.y = torch.tensor(([1],[2],[3],[4],[5],[6],[7]), detype=torch.int64)
        #self.X = torch.from_numpy(self.y)
        self.y = torch.tensor(self.y, dtype = torch.float32)
        print(self.y)
        #ensure the input data is floats
        #label enconde target and ensure the values are floats
        #self.y = LabelEncoder().fot_transform(self, y)
        #self.y = self.y.astype('float32')
        #self.y = self.y.reshape((len(self.y),1 ))
        
    #number of row in the dataset    
    def __len__(self):
        return len(self.X)
    #get a row at an index
    def __getitem__(Self,idx):
        return [self.X[idx], self.y[idx]]
    
    
#model definition    
class MLP(Module):
    #define model elements
    def __init__(self,n_inputs):
        super(MLP,self).__init__()
        #input to firt hidden layer
        self.hidden1 = Linear(n_inputs, 10)
        kaiming_uniform_(self.hidden1.weight, nonliearity='relu')
        self.act1 = ReLU()
        
        self.hidden2 = Linear(10, 10)
        kaiming_uniform_(self.hidden2.weight, nonliearity='relu')
        self.act2 = ReLU()
        
        self.hidden3 = Linear(10, 7)
        kaiming_uniform_(self.hidden3.weight)
        self.act3 = Sigmoid()
        print('Done model')
        
     #forward propagate input   
    def forward(Self, X):
        #input to first hidden layer
        X = self.hidden1(X)
        print(X)
        X = self.act1(X)
        
        #input to second hidden layer    
        X = self.hidden2(X)
        X = self.act2(X)
        
        #input to third hidden layer & output
        X = self.hidden3(X)
        X = self.act3(X)
        print('Done model')
        return X
    
#prepare dataset        
def prepare_data(path):
    #load dataset
    dataset = CSVDataset(path)
    for a in dataset:
        print(a)  
        #calculate split
        #train, test = dataset.get_splits()
        #preapare data loaders
        train_dl = DataLoader(dataset, bath_size=32, shuffle = True)
        #test_dl = DataLoader(test, batch_size=1024, shuffle=False)
        return train_dl
    
#train the model    
def train_model(train_dl, model):
    criterion = BCELoss()
    optimizer = SGD(model.parameters(), lr=0.01, momentum=0.9)
    #enumeate epochs
    for epoch in range(100):
        #enumarate mini batches
        for i,(inputs, targets) in enumerate(train_dl):
            #clear gradients
            optimizer.zero_grad()
            #compute the model inpui
            yhat = model(inputs)
            #calculate loss
            loss = criterion(yhat, targets)
            #credit assignment
            loss.backward()
            #update model weights
            optimizer.step()
            
#EVALUATE MODEL
def evaluate_model(test_dl, model):
    predictions, actuals = list(), list()
    for i, (inputs, targets) in enumerate(test_dl):
        #evaluate the model on the test set
        yhat = model(inputs)
        #retrieve numpy array
        yhat = yhat.detach().numpy
        actual = targets.numpy()
        actual = actual.reshape((len(actual),1))
        #round to class values
        yhat = yhat.round()
        #store
        predictions.append(actual)
    predictions, actuals = vstack(predictions), vstack(actuals)
    #calculate accuracy
    acc = accuracy_score(actuals, predictions)
    return acc

#make a class prediction for  one row of data
def predict(roc, model):
    #convert row to data
    row = Tensor([row])
    #make prediction
    yhat = model (row)
    print(yhat)
    #retrieve numpy array
    yhat = yhat.detach().numpy()
    return yhat    

#prepare the data
path = "Training_Dataset.csv"
train_del = prepare_data(path)
print(len(train_dl.dataset), len(test_dl.dataset))
#define the network
model= MLP(4)
#evaluate model
acc = evaluate_model(test_dl, model)
#print('Accuracy> %.3f', %acc)
#make a single prediction
#rom = [1,0,75,78]
#yhat = predict(row, model)
#print('Predicted> %.3f (class=%d)'%(yhat, yhat.round()))
```
